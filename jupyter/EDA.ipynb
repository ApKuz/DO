```python
# libraries
import sys
import pandas as pd
import numpy as np
import utm

# file read
sys.path.append('..')
import helper

with open('../Data/blueroute1export.json') as f:
    _df = pd.read_json(f)



# options
pd.options.display.max_columns = 150
```

```python
# keeping _df in cache so we do not have to reopen the data if a function does not perform properly
df = _df
```

```python
# Create a new timeStamp column that is the '$date' entry in 'timeField'
df['timeStamp'] = df['timeField'].apply(lambda x: x['$date'])

# remove 'timeStamp' column and move it to the first column
firstColumn = df.pop('timeStamp')
df.insert(0, 'timeStamp', firstColumn)

# convert 'timeStamp' column to pandas timestamp
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
x = ['timeStamp', 'timeField', 'topic', 'size', '_id', 'msg_type', 'metadataID', 'header', 'measurementTime', 'road', 'measurement', 'routingRequest', 'mapVersion', 'status', 'transforms', 'signal', 'engageAdvice', 'gearLocation', 'steeringRate']
```

```python
for value in dict:
    dict.update({value: dict[value].reset_index()})
    
    # We will drop the below fields
    for field in ['index', 'timeField', 'topic', 'size', 'msg_type', 'metadataID', '_id', 'header']:
        
        # drop iff it is in the subDataFrame
        try:
            tempDF = dict[value].drop(columns=[field])
        except KeyError:
            continue
        else:
            dict.update({value: tempDF})
    
    # sort all DataFrames by the timestamp
    dict.update({value: dict[value].sort_values(by=['timeStamp'])})
```

```python
dict['df0']
```

```python
for value in dict:
    
    if value == 'df0':
        mergedDF = dict[value]
        
    else:
        mergedDF = pd.merge_asof(mergedDF, dict[value], on=['timeStamp'])
        
mergedDF['timeStamp'] = mergedDF['timeStamp'].sort_values()
```

```python
mergedDF['transforms_x'][0][0]['transform']['translation']['x']
```

```python
mergedDF['lat'] = np.nan
mergedDF['lon'] = np.nan

for index, row in mergedDF.iterrows():
    
    # get easting and northing from transforms column
    easting = mergedDF['transforms_x'][index][0]['transform']['translation']['x']
    northing = mergedDF['transforms_x'][index][0]['transform']['translation']['y']
    
    
    # easting must follow the below parameters
    if not (easting >= 100_000 and easting <= 1_000_000):
        continue
    
    # finding the latlong with provided UTM
    latLong = utm.to_latlon(easting, northing, 17, 'N')
    
    # setting the lat and lon values in df
    mergedDF.at[index, 'lat'] = latLong[0]
    mergedDF.at[index, 'lon'] = latLong[1]
```
